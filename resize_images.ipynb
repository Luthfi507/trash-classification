Karena resize gambar memerlukan waktu yang lama sehingga dataset hasil resize disimpan ke github yang kemudian datasetnya digunakan untuk eksperimen modeling

```python
# membuat direktori untuk setiap label
for label in labels:
    os.makedirs(f'data/{label}', exist_ok=True)

# menyimpan gambar sesuai dengan label direktori
for i in range(len(ds['train'])):
    label_id = ds['train'][i]['label'] # mengambil label id gambar
    label = id2label[label_id]

    image = ds['train'][i]['image'] # mengambil gambar
    image = image.resize((224, 224)) # mengubah ukuran gambar menjadi (224, 224)

    filename = f'{label}_{i}.jpg'
    image.save(f'data/{label}/{filename}') # menyimpan gambar
```
